Questão 01: (2,0 pontos)

Dada a equação ```f(x) = 3x4 - 2x3 - x2 + x +8 = 0```, obtenha o valor aproximado da raiz. Utilizando o método da bissecção no intervalo

 ```[-1; 1]``` com precisão ```1E-06```.

Absoluto:
```
i. abs(f(x))<= 1E-06
ii. abs(f(b)-f(a)) <= 1E-06
iii. abs(b-a) <= 1E-06
```

Relativo:
```
i. abs((f(b)-f(a))/f(b)) <= 1E-06
ii. abs((b-a)/b) <= 1E-06
```

````python
import math

a = -1
b = 1
nivelPrecisao = 0.000001

def f (xo):
   return (3*pow(xo, 4)) - (2*pow(xo, 3)) - pow(xo, 2) + xo + 8

if (b - a ) < nivelPrecisao:
   xb = a
else:
   k = 1
   M = f(a)
   while True:
      x = (a + b)/2
      print('---------------')
      print('Interação: ', k)
      print('Valor de X: ', x)
      print('f(x): ', f(x))
      print('b - a: ', b - a/2)
      print('---------------','\n')
      if M * f(x) > 0:
         a = x
      else:
         b = x
      if (b - a) <= nivelPrecisao:
         xb = (a + b)/2
         break
      k = k + 1

print('[-1;1] = ~', xb,'\n')
````

Sejam ```x=[1; 3; -5; 6; 6]``` e sua aproximação ```x̄=[2; -4; 6; 8; -7]```. 

Calcule o erro absoluto e o erro relativo (usar x como referência).

````python
import math

x = [1, 3, -5, 6, 6]
xb = [2, -4, 6, 8, -7]
print('Sendo X: ', x, '\n e\nXBarra:', xb,'\n')

def n2(Vet):
    i=0
    xw =[]
    while i < len(Vet):
       xw.append(Vet[i]**2)
       print(xw)
       i = i+1
    i=0
    valor = 0
    while i < len(xw):
      valor = valor + xw[i]
      i=i+1
    print(('\nRaiz² de {}'.format(valor)),'\n')
    return math.sqrt(valor)

print('-------------------\n')

def erroAbsoluto(Vet1,Vet2):
    y=0
    valor = []
    while y < len(Vet1):
      valor.append(Vet1[y]-Vet2[y])
      y = y+1
      print(valor)
    return n2(valor)

def erroRelativo(Vet1,Vet2,erroAbsoluto):
    y=0
    valor = []
    while y < len(Vet1):
      valor.append(Vet1[y]-Vet2[y])
      y = y+1  
    valor = erroAbsoluto / n2(Vet1)
    print(valor)
    return valor

erro_absoluto = erroAbsoluto(x,xb)

print('\nErro Absoluto: {}'.format(erro_absoluto),'\n')
print('-------------------\n')
print('\nErro Relativo {}'.format(erroRelativo(x,xb,erro_absoluto)))
````
